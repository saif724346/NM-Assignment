# **3. Perform the Below Visualizations.**

# **Univariate Analysis:**
Univariate analysis involves exploring the distribution of a single variable. We can use histograms, box plots, and other plots to visualize the distribution of the data.

```python
sns.distplot(df['Distance from the airport'])
plt.show()
```

# **Univariate Analysis:**
Univariate analysis involves exploring the distribution of a single variable. We can use histograms, box plots, and other plots to visualize the distribution of the data.

```python
sns.boxplot(df['condition of the house'])
plt.show()
```

```python
#mean
mean_price = df['Price'].mean()
print("Average House Price:", mean_price)
```

```python
mode_floor = df['number of floors'].mode()
print("maximum number of house in floor",mode_floor)
```

# **Bi-Variate Analysis:**
Bivariate analysis involves exploring the relationship between two variables. We can use scatter plots, heatmaps, and other plots to visualize the relationship between the two variables.

```python
sns.scatterplot(x='number of bedrooms', y='Price', data=df)
plt.show()
```

```python
sns.jointplot(x='Distance from the airport', y='Price', data=df)
plt.show()
```

# **Multi-Variate Analysis:**
Multivariate analysis involves exploring the relationship between more than two variables. We can use scatter plots, heatmaps, and other plots to visualize the relationship between multiple variables.

```python
sns.pairplot(df[['Number of schools nearby', 'number of bedrooms','Distance from the airport', 'Price']])
plt.show()
```

```python
# Multiple linear regression
from sklearn.linear_model import LinearRegression
X = df[['Lattitude', 'Longitude', 'Price']]
y = df['Price']
reg = LinearRegression()
reg.fit(X, y)
y_pred = reg.predict(X)
plt.scatter(y, y_pred)
plt.show()
```

# **5. Handle the Missing values.**
Handling missing values in Python can be done using various libraries such as Pandas or NumPy. Here's an example of how to handle missing values using Pandas:

```python
# Check for missing values
print(df.isnull().sum())

# Option 1: Remove rows with missing values
data1 = df.dropna()

# Option 2: Impute missing values with mean
mean = df.mean()
data2 = df.fillna(mean)

# Option 3: Impute missing values with interpolation
data3 = df.interpolate()

# Print the first 5 rows of each dataset to compare
print(data1.head())
print(data2.head())
print(data3.head())
```
